```python
import sys
import os

# Menambahkan folder root proyek ke sys.path
sys.path.append(os.path.abspath(".."))

from simulated_annealing import SimulatedAnnealing
from cube.visual import plot_cube
```

```python
main = SimulatedAnnealing(initial_temperature=1000, cooling_rate=0.99, temperature_threshold=0.001, stagnation_threshold=50)
initial_state = main.initial_state
initial_score = main.objective_function
```

```python
plot_cube(final_state, f"Final State of the Cube (Score: {final_score}, Time: {execution_time:.2f}s, Total Stuck: {main.total_stuck_count})")
```
