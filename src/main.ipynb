1. Steepest Ascendt Hill Climbing

```python
from algorithms.SteepestHillClimb import SteepestHillClimbing
from cube.visual import plot_cube
import matplotlib.pyplot as plt
```

```python
main = SteepestHillClimbing(max_iterations=100)
initial_state = main.initial_state
initial_score = main.objective_function


# Visualisasi kondisi awal
plot_cube(initial_state, f"Initial State of the Cube (Score: {initial_score})")
```

```python
final_state, final_score, execution_time, history, iteration_count = main.search()
```

```python
# Visualisasi kondisi akhir
plot_cube(final_state, f"Final State of the Cube (Score: {final_score}, Time: {execution_time:.2f} seconds, Iteration: {iteration_count-1})")
```

```python
# Visualisasi skor tiap iterasi
plt.plot(history)
plt.xlabel('Iteration')
plt.ylabel('Score')
plt.title('Score per Iteration')
plt.show()
```
